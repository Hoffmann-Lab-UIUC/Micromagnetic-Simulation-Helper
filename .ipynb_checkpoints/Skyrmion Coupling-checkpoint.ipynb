# In plane magnetic field

```python
# NUMERICAL PARAMETERS RELEVANT FOR THE SPECTRUM ANALYSIS
fmax = 30e9        # maximum frequency (in Hz) of the sinc pulse
T    = 20e-9        # simulation time (longer -> better frequency resolution)
dt = 1/(2*fmax)  # the sample time (Nyquist theorem taken into account)
```

```python
# Note that this is a format string, this means that the statements inside the
# curly brackets get evaluated by python. In this way, we insert the values of
# the variables above in the script.

def resonance_script():

    script = f"""
    nx   := 128
    setgridsize(300,300,1)
    setcellsize(1.5e-9,1.5e-9,1.5e-9)

    domain := circle(400e-9)

    setgeom(domain)

    Msat = 1e6
    Aex = 20e-12
    Dind = 3e-3
    Ku1 = 10e5
    AnisU = vector(0,0,1)
    alpha = 0.03

    m = NeelSkyrmion(1,-1)

    B_ext = vector(.01 * sinc(2*pi*{fmax}*t),0,0)

    // dia := 80e-9
    // shape_1 := circle(dia).transl(-150e-9,0,0)
    // shape_2 := circle(dia)
    // shape_3 := circle(dia).transl(150e-9,0,0)

    // m = Uniform(0, 0, 1)
    // m.setInShape(shape_1, NeelSkyrmion(1, -1).transl(-120e-9,0, 0))
    // m.setInShape(shape_2, NeelSkyrmion(1, -1))
    // m.setInShape(shape_3, NeelSkyrmion(1, -1).transl(120e-9,0,0))

    minimize()
    autosave(m,{dt})
    tableautosave({dt})
    run({T})
    """


    return script

run_mumax3(resonance_script(),"In plane field excite")
```

```python
fmax = 30e9

table = read_mumax3_table(r"C:\Users\foile\OneDrive - University of Illinois - Urbana\Research\Calculations\In plane field excite.out\table.txt")

plt.figure(figsize=(10, 8))
dm = table["mx"] - np.median(table["mx"])
spectr = np.abs(np.fft.fft(dm))
real_spectr = spectr[:int(np.ceil(len(spectr)/2))]
freq = np.linspace(0, fmax, len(real_spectr))


plt.plot(freq/1e9, real_spectr)
plt.xlim(0, 30)
plt.xlabel('Frequency (GHz)')
plt.ylabel('Spectral Power')
plt.legend()
plt.savefig("AnistropyResonance.png",facecolor='white')
plt.show()

```
